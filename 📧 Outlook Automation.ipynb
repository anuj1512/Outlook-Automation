# ðŸ‘†Case 1- When we want to download the attachments from a particular mail address on a particular date

```python
 #pip install pypiwin32 to work with windows operating sysytm
import win32com.client
import datetime
import os
from datetime import date
import time
t1=time.time()
print("Execution of the script started")

# Creating an object for the outlook application.

outlook = win32com.client.Dispatch("Outlook.Application").GetNamespace("MAPI")

# Creating an object to access Inbox of the outlook.

inbox=outlook.GetDefaultFolder(6)


date='2022-04-21'
# creating a folder with date as its name where we will store our extracted attachments 
fp=r"C:\Users\euakumn\OneDrive - Ericsson AB\Desktop\Outlook_attachment"+"\\"+date
os.mkdir(fp)

messages=inbox.Items
count=0
for i in messages:
   
   
    try:
        # In below line we can define the email address of which we want to download all the attachment on a particular date        
        if i.SenderEmailAddress=='upm@airtel.com' and str(i.senton.date())==date:
            attachments=i.Attachments
            attachment = attachments.Item(1)
            # the name of attachment file      
            attachment_name = str(attachment)
            # Below line saves the attachments in that mail to a particular location which we can define             
            attachment.SaveASFile(r"C:\Users\euakumn\OneDrive - Ericsson AB\Desktop\Outlook_attachment"+"\\"+date+"\\"+attachment_name)
            print(i.SenderEmailAddress,i.senton.date())
            count+=1 
        
    except AttributeError as e:
        pass
print(count,"files successfully extracted...!!!")
```

# ðŸŽ¯Case 2- If we want to extract the attachments between 2 given dates with various search customization like senders name, sender's email address or by any keyword

```python
#pip install pypiwin32 to work with windows operating sysytm
import win32com.client 
import datetime
import os
import re
import time
t1=time.time()
print("Execution started....Pls wait")

# Creating an object for the outlook application.
outlook = win32com.client.Dispatch("Outlook.Application").GetNamespace("MAPI")

# Creating an object to access Inbox of the outlook.
inbox=outlook.GetDefaultFolder(6)


date='2021-07-24'
fp=r"C:\Users\euakumn\OneDrive - Ericsson AB\Desktop\Outlook_attachment"+"\\"+date
#  Creating a folder with name as date where we will store extracted files
os.mkdir(fp)


messages=inbox.Items
# Highdate is the upper limit of the data with yy-mm-dd format
Highdate=datetime.date(2022,4,9)
# Lowdate is the lower limit of the data with yy-mm-dd format
Lowdate=datetime.date(2022,4,7)

# We will search for the attachment from a particular mail address/Mail subject between these 2 dates

count=0
for i in messages:
    try:
        m=i.senton.date()
        # Here we are looking for a keyword "R&P_Dashboard" in mail subject      
        if (m>=Lowdate and m<=Highdate) and re.search('R&P_Dashboard',i.subject)!=None and  i.Sender==None :
            print(i,i.sender,i.senton.date())
            attachments=i.Attachments
            attachment = attachments.Item(1)
            # the name of attachment file      
            attachment_name = str(attachment)
            # Below logic will save the attachments from all these mails to a particular folder which we have already defined             
            attachment.SaveASFile(r"C:\Users\euakumn\OneDrive - Ericsson AB\Desktop\Outlook_attachment"+"\\"+date+"\\"+attachment_name+date)
            print(i,i.sender,i.senton.date())
            count=count+1
        else:
            pass
    except  AttributeError as e:
        pass
        
print('Total no. of result: ',count)
print(f'Successfully completed in {time.time()-t1} seconds')
```

# Case 4 - Moving mail from a particular email address/subject to another folder automatically

```python
#pip install pypiwin32 to work with windows operating sysytm
import win32com.client 
import datetime
import os
import re
import time
t1=time.time()
print("Execution started....Pls wait")

# Creating an object for the outlook application.
outlook = win32com.client.Dispatch("Outlook.Application").GetNamespace("MAPI")

# Creating an object to access Inbox of the outlook.
inbox=outlook.GetDefaultFolder(6)

messages=inbox.Items

# Highdate is the upper limit of the data with yy-mm-dd format
Highdate=datetime.date(2022,4,1)
# Lowdate is the lower limit of the data with yy-mm-dd format
Lowdate=datetime.date(2022,4,1)

count=0
# We have mannully created a blank folder inside inbox in Outlook with name as "RNP Mails"
RNP=inbox.folders['RNP Mails']

#  Lets assume we are getting mail having "R&P" keyword in mail's subject and we want to move these mails from inbox to a new folder
""" We will search for the particular mail address/Mail subject between these 2 dates
and move those mails into "RNP Mails" folders"""

for i in messages:
    try:
        m=i.senton.date()
                 
        if (m>=Lowdate and m<=Highdate) and re.search('R&P',i.subject)!=None and i.sender==None:
            print(i,i.sender,i.senton.date())
            i.move(RNP)
#
            count=count+1
        else:
            pass
    except  AttributeError as e:
        pass
        
print('Total no.Emails moved from Inbox to RNP Mails folder: ',count)
print(f'Script Successfully completed in {time.time()-t1} seconds')
```
